```python
def get_2_max(lst):
  if len(lst) < 2:
    raise ValueError('List must have at least 2 elements')
  lrgst1 = lst[0]
  lrgst2 = float('-inf')
  for i in range(1, len(lst)):
    if lst[i] > lrgst1:
      lrgst2 = lrgst1
      lrgst1 = lst[i]
    elif lrgst2 < lst[i] < lrgst1:
      lrgst2 = lst[i]
  if lrgst2 == float('inf'):
    raise ValueError('List must have at least 2 distinct elements')
  return [lrgst1, lrgst2]
```

```python
lst = [2,2,2,2,2,2,2]
print(lst)
print(get_2_max(lst))
```

```python
def get_second_min(lst):
  if len(lst) < 2:
    raise ValueError('List must have at least 2 elements')
  min1 = min2 = float('inf')
  for i in range(1, len(lst)):
    if lst[i] < min1:
      min2 = min1
      min1 = lst[i]
    elif min1 < lst[i] < min2:
      min2 = lst[i]
  if min2 == float('inf'):
    raise ValueError('List must have at least 2 distinct elements')
  return min2
```
